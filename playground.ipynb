```python
import os
import csv
import pandas as pd

from auto_glosh import find_best_min_pts
```

```python
CURRENT_PATH = os.getcwd()
```

```python
def calc_min_pts():
    files = [["Path", "mpts_star"]]
    errors = []
    with open("metadata.csv", 'r') as f:
        reader = csv.DictReader(f)
        for row in reader:
            file_path = os.path.join(row["Path"] + "/" + row["Name"])
            outlier_path = file_path.replace(".csv", "_outliers.csv")
            glosh_profiles = pd.read_csv(outlier_path)
            try:
                mpts_star = find_best_min_pts(glosh_profiles)
                files.append([file_path, mpts_star])
            except Exception as err:
                errors.append(file_path)

    with open("output/best_mpts.csv", "w") as f:
        writer = csv.writer(f)
        writer.writerows(files)
    
    with open("output/errors.csv", "w") as f:
        writer = csv.writer(f)
        writer.writerows(errors)
```
